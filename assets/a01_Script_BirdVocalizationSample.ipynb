```python
# From folder ../Audios/ create one folder per subfolder, with the same name in ../Bird Vocalization Samples and copy the shortest audio file from each subfolder to the corresponding folder in ../Bird Vocalization Samples
import os
import shutil
import librosa
import numpy as np
import matplotlib.pyplot as plt
import librosa.display
from PIL import Image
from io import BytesIO

# Get the list of subfolders in ../Audios/
subfolders = [f.path for f in os.scandir('../Audios/') if f.is_dir()]

# For each subfolder
for subfolder in subfolders:
    # Get the name of the subfolder
    subfolder_name = subfolder.split('/')[-1]
    # Create a folder with the same name in ../Bird Vocalization Samples
    os.makedirs('../Bird Vocalization Samples/' + subfolder_name, exist_ok=True)
    # Get the list of audio files in the subfolder if they are .WAV files
    audio_files = [f.path for f in os.scandir(subfolder) if f.is_file() and f.path.endswith('.WAV')]

    # Get the shortest audio file
    shortest_audio_file = min(audio_files, key=os.path.getsize)

    output_filepath = shortest_audio_file.replace('../Audios/', '../Bird Vocalization Samples/')
    output_path = output_filepath.replace(shortest_audio_file.split('/')[-1], '')
    # Copy the shortest audio file to the corresponding folder in ../Bird Vocalization Samples
    print('Copying ' + shortest_audio_file + ' to ' + output_path)
    shutil.copy(shortest_audio_file, output_path)

    # Generate mel spectrogram
    fmin = 1
    fmax = 32000
    y, sr = librosa.load(output_filepath, sr=None)
    mel_spectrogram = librosa.feature.melspectrogram(y=y, sr=sr)
    mel_spectrogram_path = output_filepath.replace('.WAV', '.PNG')
    fig, ax = plt.subplots(figsize=(12, 6))
    D = librosa.amplitude_to_db(librosa.stft(y), ref=np.max)
    librosa.display.specshow(D, sr=sr, x_axis="time", y_axis="log", fmin=fmin, fmax=fmax, ax=ax)
    ax.axis('off')
    buf = BytesIO()
    fig.savefig(buf, format='png', bbox_inches='tight', pad_inches=0)
    plt.close(fig)
    buf.seek(0)
    image = Image.open(buf)
    image.save(mel_spectrogram_path)    

print('Done!')
```
